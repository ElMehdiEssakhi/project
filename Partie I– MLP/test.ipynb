```python
dataset_2 = dataset["train"].train_test_split(test_size=0.2, seed=42)
```

```python
train = dataset_2["train"].to_pandas()
test = dataset_2["test"].to_pandas()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
education_order = {
    "Preschool": 0,
    "1st-4th": 1,
    "5th-6th": 2,
    "7th-8th": 3,
    "9th": 4,
    "10th": 5,
    "11th": 6,
    "12th": 7,
    "HS-grad": 8,
    "Some-college": 9,
    "Assoc-voc": 10,
    "Assoc-acdm": 11,
    "Bachelors": 12,
    "Masters": 13,
    "Prof-school": 14,
    "Doctorate": 15
}

train["education"] = train["education"].map(education_order)
test["education"] = test["education"].map(education_order)
```

```python
train["income"] = train["income"].map({
    "<=50K": 0,
    ">50K": 1
})
```

```python
test["income"] = test["income"].map({
    "<=50K": 0,
    ">50K": 1
})
```

```python
freq = train["native.country"].value_counts()
rare = freq[freq < 100].index
```

```python
train["native.country"] = train["native.country"].replace(rare, "Other")
```

```python
test["native.country"] = test["native.country"].apply(
    lambda x:x if x in freq.index and x not in rare else "Other")
```

```python
train["native.country"].value_counts()
```

```python
test["native.country"].value_counts()
```

```python
ohe_cols = [
    "workclass", "marital.status", "occupation",
    "relationship", "race", "sex", "native.country"
]
```

```python
ohe = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
ohe.fit(train[ohe_cols])
```

```python
train_encoded = pd.DataFrame(
    ohe.transform(train[ohe_cols]),
    columns=ohe.get_feature_names_out(ohe_cols),
    index=train.index
)
test_encoded = pd.DataFrame(
    ohe.transform(test[ohe_cols]),
    columns=ohe.get_feature_names_out(ohe_cols),
    index=test.index
)
```

```python
train.drop(columns=ohe_cols, inplace=True)
test.drop(columns=ohe_cols, inplace=True)
```

```python
train =pd.concat([train, train_encoded], axis=1)
test = pd.concat([test, test_encoded], axis=1)
```

```python
train.head()
```

```python
num_cols = [
    "age", "fnlwgt", "capital.gain",
    "capital.loss", "hours.per.week", "education.num", "education"
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(train[num_cols])
```

```python
train[num_cols] = scaler.transform(train[num_cols])
test[num_cols] = scaler.transform(test[num_cols])
```
